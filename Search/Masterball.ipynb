### a) Create a class to the model the Masterball problem.


```python
import util
import search
import sys
import copy
import random
from collections import deque

def turn(state, n):
    r = list()
    for i in xrange(4):
        aux = deque(state[i])
        aux.rotate(-n)
        lista = list(aux)
        list1 = lista[:4]
        list2 = lista[4:]
        list1.reverse()
        aux = deque(list1+list2)
        aux.rotate(n)
        r.append(list(aux))
    return r

class MasterBallProblem(search.SearchProblem):
    
    def __init__(self,startState):
        self.start = list(startState)
        self.expanded = 0
    
    def getStartState(self):
        return self.start
    
    def isGoalState(self,state):
        for i in xrange(1,8):
            if ((state[0][i-1])+1) % 8 == (state[0][i]) % 8:
                for j in xrange(1,4):
                    if state[j-1][i-1] != state[j][i-1]:
                        return False
            else:
                return False
            
        return True

    def getSuccessors(self,state):
        
        R_0 = "Right 0"
        R_1 = "Right 1"
        R_2 = "Right 2"
        R_3 = "Right 3"
        
        L_0 = "Left 0"
        L_1 = "Left 1"
        L_2 = "Left 2"
        L_3 = "Left 3"

        V_0 = "Vertical 0"
        V_1 = "Vertical 1"
        V_2 = "Vertical 2"
        V_3 = "Vertical 3"
        V_4 = "Vertical 4"
        V_5 = "Vertical 5"
        V_6 = "Vertical 6"
        V_7 = "Vertical 7"
        
        successors = []
        
        #R_0
        r0 = list(state)
        r0[0] = state[0][7:]+state[0][:7]
        successors.append((r0,R_0))
        
        #R_1
        r1 = list(state)
        r1[1] = state[1][7:]+state[1][:7]
        successors = successors + [(r1,R_1)]
        
        #R_2
        r2 = list(state)
        r2[2] = state[2][7:]+state[2][:7]
        successors = successors + [(r2,R_2)]
        
        #R_3
        r3 = list(state)
        r3[3] = state[3][7:]+state[3][:7] 
        successors = successors + [(r3,R_3)]
        
        #V_0
        successors = successors + [(turn(state,0),V_0)]
        
        #V_1
        successors = successors + [(turn(state,1),V_1)]
        
        #V_2
        successors = successors + [(turn(state,2),V_2)]
        
        #V_3
        successors = successors + [(turn(state,3),V_3)]
        
        #V_4
        successors = successors + [(turn(state,4),V_4)]
        
        #V_5
        successors = successors + [(turn(state,5),V_5)]
        
        #V_6
        successors = successors + [(turn(state,6),V_6)]
        
        #V_7
        successors = successors + [(turn(state,7),V_7)]
        
        self.expanded += 8
        return successors
    
    def randomSuccessor(self,state):
        
        r = random.randint(0, 11)
        
        L_0 = "Left 0"
        L_1 = "Left 1"
        L_2 = "Left 2"
        L_3 = "Left 3"

        V_0 = "Vertical 0"
        V_1 = "Vertical 1"
        V_2 = "Vertical 2"
        V_3 = "Vertical 3"
        V_4 = "Vertical 4"
        V_5 = "Vertical 5"
        V_6 = "Vertical 6"
        V_7 = "Vertical 7"
        
        if r == 0:
            #L_0
            l0 = list(state)
            l0[0] = state[0][1:]+state[0][:1]
            return L_0, l0
        
        if r == 1:
            #L_1
            l1 = list(state)
            l1[1] = state[1][1:]+state[1][:1]
            return L_1, l1
        
        if r == 2:
            #L_2
            l2 = list(state)
            l2[2] = state[2][1:]+state[2][:1]
            return L_2, l2
        
        if r == 3:
            #L_3
            l3 = list(state)
            l3[3] = state[3][1:]+state[3][:1]
            return L_3, l3    
        
        if r == 4:
            #V_0
            return V_0, turn(state, 0)
        
        if r == 5:
            #V_1
            return V_1, turn(state, 1)
        
        if r == 6:
            #V_2
            return V_2, turn(state, 2)
        
        if r == 7:
            #V_3
            return V_3, turn(state, 3)
        
        if r == 8:
            #V_4
            return V_4, turn(state, 4)
        
        if r == 9:
            #V_5
            return V_5, turn(state, 5)
        
        if r == 10:
            #V_6
            return V_6, turn(state, 6)
        
        if r == 11:
            #V_7
            return V_7, turn(state, 7)
```

### b) Implement iterative deepening search and A* search.

```python
def depthLimitedSearch(problem,limit):
    visited = {}
    counter = 0
    state = problem.getStartState()
    frontier = util.Stack()
    s = str(state)
    depth = {}
    depth[s] = 0
    frontier.push((state, ["Start"]))
    visited[s] = 'gray'
    while not frontier.isEmpty():
        u, actions = frontier.pop()
        counter += 1
        uprime = str(u)
        #print uprime,actions
        if problem.isGoalState(u):
            return  actions, counter
        if depth[uprime] == limit:
            continue
        for v, action in problem.getSuccessors(u):
            vprime = str(v)
            if not vprime in visited:
                visited[vprime] = 'gray'
                depth[vprime] = depth[uprime] + 1
                frontier.push((v, actions + [action]))
        visited[uprime] = 'black'
    return [], 0

def IDS(problem):
    for i in xrange(sys.maxint):
        result, nodes = depthLimitedSearch(problem,i)
        if len(result) > 0:
            return result, nodes
        
def aStarSearch(problem, heuristic):
    visited = {}
    movs = {}
    counter = 0
    pqueue = util.PriorityQueue()
    state = problem.getStartState()
    
    visited[str(state)] = 1
    movs[str(state)] = ["Start"]
    pqueue.push((state,"Start"), heuristic(state))
    
    while not pqueue.isEmpty():
        counter +=1
        state, mov = pqueue.pop()
        
        if problem.isGoalState(state):
            #print "Goal \n", visited
            return movs[str(state)], counter 
        else:
            for st, mov in problem.getSuccessors(state):
                    if str(st) not in visited: 
                        visited[str(st)] = visited[str(state)] + 1
                        movs[str(st)] = list(movs[str(state)])
                        movs[str(st)].append(mov)
                        pqueue.push((st,mov), (visited[str(st)] + heuristic(st)))
                        #print heuristic(st), visited[str(st)], st, mov

    #print "No encontrado"
    return [], 0
```

### Effective Branching Factor
The effective branching factor is giving by:

$$N+1 = 1 + {b}^{*} + {{b}^{*}}^{2} + ... + {{b}^{*}}^{d}$$


$$N+1 = \frac{{{b}^{*}}^{d+1}}{{b}^{*}+1} $$

Where N is the number of nodes expanded, d is the depth of the search and b* is the branch factor.

### a) Create a class to the model the Masterball problem.


```python
import util
import search
import sys
import copy
import random

class MasterBallProblem(search.SearchProblem):
    
    def __init__(self,startState):
        self.start = list(startState)
        self.expanded = 0
    
    def getStartState(self):
        return self.start
    
    def isGoalState(self,state):
        for i in range(1,4):
            for j in range(8):
                if state[i][j] != state[i-1][j]:
                    return False
        return True
    
    def getSuccessors(self,state):
        
        R_0 = "Right 0"
        R_1 = "Right 1"
        R_2 = "Right 2"
        R_3 = "Right 3"
        
        L_0 = "Left 0"
        L_1 = "Left 1"
        L_2 = "Left 2"
        L_3 = "Left 3"

        V_0 = "Vertical 0"
        V_1 = "Vertical 1"
        V_2 = "Vertical 2"
        V_3 = "Vertical 3"
        V_4 = "Vertical 4"
        V_5 = "Vertical 5"
        V_6 = "Vertical 6"
        V_7 = "Vertical 7"
        
        successors = []
        
        #R_0
        r0 = list(state)
        r0[0] = state[0][7:]+state[0][:7]
        successors.append((r0,R_0))
        
        #R_1
        r1 = list(state)
        r1[1] = state[1][7:]+state[1][:7]
        successors = successors + [(r1,R_1)]
        
        #R_2
        r2 = list(state)
        r2[2] = state[2][7:]+state[2][:7]
        successors = successors + [(r2,R_2)]
        
        #R_3
        r3 = list(state)
        r3[3] = state[3][7:]+state[3][:7] 
        successors = successors + [(r3,R_3)]
        
        """
        #L_0
        l0 = list(state)
        l0[0] = state[0][1:]+state[0][:1]
        successors.append((l0,L_0))
        
        #L_1
        l1 = list(state)
        l1[1] = state[1][1:]+state[1][:1]
        successors.append((l1,L_1))
        
        #L_2
        l2 = list(state)
        l2[2] = state[2][1:]+state[2][:1]
        successors.append((l2,L_2))
        
        #L_3
        l3 = list(state)
        l3[3] = state[3][1:]+state[3][:1]
        successors.append((l3,L_3))
        """
        
        #V_0
        v0 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(4):
                (v0[i][j],v0[3-i][3-j]) = (v0[3-i][3-j],v0[i][j])
        successors = successors + [(v0,V_0)]
        
        #V_1
        v1 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(1,5):
                (v1[i][j],v1[3-i][4-j]) = (v1[3-i][4-j],v1[i][j])
        successors = successors + [(v1,V_1)]
        
        #V_2
        v2 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(2,6):
                (v2[i][j],v2[3-i][5-j]) = (v2[3-i][5-j],v2[i][j])
        successors = successors + [(v2,V_2)]
        
        #V_3
        v3 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(3,7):
                (v3[i][j],v3[3-i][6-j]) = (v3[3-i][6-j],v3[i][j])
        successors = successors + [(v3,V_3)]
        
        #V_4
        v4 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(4,8):
                (v4[i][j],v4[3-i][7-j]) = (v4[3-i][7-j],v4[i][j])
        successors = successors + [(v4,V_4)]
        
        #V_5
        v5 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(5,9):
                (v5[i][j%8],v5[3-i][(8-j)%8]) = (v5[3-i][(8-j)%8],v5[i][j%8])
        successors = successors + [(v5,V_5)]
        
        #V_6
        v6 = copy.deepcopy(state)
        for i in xrange(2):
            for j in xrange(6,10):
                (v6[i][j%8],v6[3-i][(9-j)%8]) = (v6[3-i][(9-j)%8],v6[i][j%8])
        successors = successors + [(v6,V_6)]
        
        #V_7
        v7 = copy.deepcopy(state)
        for i in range(2):
            for j in range(7,11):
                (v7[i][j%8],v7[3-i][(10-j)%8]) = (v7[3-i][(10-j)%8],v7[i][j%8])
        successors = successors + [(v7,V_7)]
        
        self.expanded += 8
        return successors
    
    def randomSuccessor(self,state):
        
        r = random.randint(0, 7)
        
        L_0 = "Left 0"
        L_1 = "Left 1"
        L_2 = "Left 2"
        L_3 = "Left 3"

        V_0 = "Vertical 0"
        V_1 = "Vertical 1"
        V_2 = "Vertical 2"
        V_3 = "Vertical 3"
        
        if r == 0:
            #L_0
            l0 = list(state)
            l0[0] = state[0][1:]+state[0][:1]
            return L_0, l0
        
        if r == 1:
            #L_1
            l1 = list(state)
            l1[1] = state[1][1:]+state[1][:1]
            return L_1, l1
        
        if r == 2:
            #L_2
            l2 = list(state)
            l2[2] = state[2][1:]+state[2][:1]
            return L_2, l2
        
        if r == 3:
            #L_3
            l3 = list(state)
            l3[3] = state[3][1:]+state[3][:1]
            return L_3, l3
        
        if r == 4:
            #V_0
            v0 = copy.deepcopy(state)
            for i in xrange(2):
                for j in xrange(4):
                    (v0[i][j],v0[3-i][3-j]) = (v0[3-i][3-j],v0[i][j])
            return V_0, v0
        
        if r == 5:
            #V_1
            v1 = copy.deepcopy(state)
            for i in xrange(2):
                for j in xrange(1,5):
                    (v1[i][j],v1[3-i][4-j]) = (v1[3-i][4-j],v1[i][j])
            return V_1, v1
        
        if r == 6:
            #V_2
            v2 = copy.deepcopy(state)
            for i in xrange(2):
                for j in xrange(2,6):
                    (v2[i][j],v2[3-i][5-j]) = (v2[3-i][5-j],v2[i][j])
            return V_2, v2
        
        if r == 7:
            #V_3
            v3 = copy.deepcopy(state)
            for i in xrange(2):
                for j in xrange(3,7):
                    (v3[i][j],v3[3-i][6-j]) = (v3[3-i][6-j],v3[i][j])
            return V_3, v3
```

### b) Implement iterative deepening search and A* search.

```python
def depthLimitedSearch(problem,limit):
    visited = {}
    counter = 0
    state = problem.getStartState()
    frontier = util.Stack()
    s = str(state)
    depth = {}
    depth[s] = 0
    frontier.push((state, ["Start"]))
    visited[s] = 'gray'
    while not frontier.isEmpty():
        u, actions = frontier.pop()
        counter += 1
        uprime = str(u)
        #print uprime,actions
        if problem.isGoalState(u):
            return  actions, counter
        if depth[uprime] == limit:
            continue
        for v, action in problem.getSuccessors(u):
            vprime = str(v)
            if not vprime in visited:
                visited[vprime] = 'gray'
                depth[vprime] = depth[uprime] + 1
                frontier.push((v, actions + [action]))
        visited[uprime] = 'black'
    return [], 0

def IDS(problem):
    for i in xrange(sys.maxint):
        result, nodes = depthLimitedSearch(problem,i)
        if len(result) > 0:
            return result, nodes
        
def aStarSearch(problem, heuristic):
    visited = {}
    movs = {}
    counter = 0
    pqueue = util.PriorityQueue()
    state = problem.getStartState()
    
    visited[str(state)] = 1
    movs[str(state)] = ["Start"]
    pqueue.push((state,"Start"), heuristic(state))
    
    while not pqueue.isEmpty():
        counter +=1
        state, mov = pqueue.pop()
        
        if problem.isGoalState(state):
            #print "Goal \n", visited
            return movs[str(state)], counter 
        else:
            for st, mov in problem.getSuccessors(state):
                    if str(st) not in visited: 
                        visited[str(st)] = visited[str(state)] + 1
                        movs[str(st)] = list(movs[str(state)])
                        movs[str(st)].append(mov)
                        pqueue.push((st,mov), (visited[str(st)] + heuristic(st)))
                        #print heuristic(st), visited[str(st)], st, mov

    #print "No encontrado"
    return [], 0
```

```python
def f(n):
    if n <= 4:
        return n
    else:
        return (7 * n) % 8

def myHeuristic(state):
    pos = {}
    for i in xrange(len(state)):
        for j in xrange(len(state[i])):
            if state[i][j] not in pos:
                pos[state[i][j]] = list()
                pos[state[i][j]].append((i,j))
            else:
                pos[state[i][j]].append((i,j))
    
    maxi = 0
    for i in pos:
        for j in xrange(len(pos[i])):
            a1, b1 = pos[i][j]
            for k in xrange(j+1, len(pos[i])):
                a2, b2 = pos[i][k]
                if a1 == a2:
                    if maxi < (f(abs(b1-b2)))+1:
                        maxi = (f(abs(b1-b2)))+1
                else:
                    if maxi < f(abs(b1-b2)):
                        maxi = f(abs(b1-b2))
    return maxi

def myHeuristic2(state):
    pos = {}
    for i in xrange(len(state)):
        for j in xrange(len(state[i])):
            if state[i][j] not in pos:
                pos[state[i][j]] = list()
                pos[state[i][j]].append((i,j))
            else:
                pos[state[i][j]].append((i,j))
    
    suma = 0
    for i in pos:
        for j in xrange(len(pos[i])):
            a1, b1 = pos[i][j]
            for k in xrange(j+1, len(pos[i])):
                a2, b2 = pos[i][k]
                if a1 == a2:
                    suma += abs(b1-b2) + 5
                else:
                    suma += abs(b1-b2)
    return suma/60.0

print myHeuristic([[3, 2, 1, 0, 6, 7, 0, 1], [0, 3, 2, 1, 0, 5, 6, 7], [4, 3, 2, 1, 4, 5, 6, 7], [5, 4, 3, 2, 4, 5, 6, 7]])
print myHeuristic2([[3, 2, 1, 0, 6, 7, 0, 1], [0, 3, 2, 1, 0, 5, 6, 7], [4, 3, 2, 1, 4, 5, 6, 7], [5, 4, 3, 2, 4, 5, 6, 7]])
```
